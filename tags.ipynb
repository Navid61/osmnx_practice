```python
import osmnx as ox

# Define popular OSM tags as a dictionary for querying geometries
tags = {
    'building': True,  # Fetch all buildings
    'amenity': ['restaurant', 'school', 'hospital', 'bank', 'parking', 'fuel'],  # Popular amenities
    'landuse': ['residential', 'commercial', 'industrial', 'retail'],  # Common land use
    'highway': ['motorway', 'primary', 'pedestrian', 'cycleway'],  # Main road types
    'natural': ['water', 'tree', 'wood'],  # Natural features
    'shop': ['supermarket', 'bakery', 'clothes', 'hardware'],  # Shop categories
    'leisure': ['park', 'playground', 'swimming_pool', 'sports_centre'],  # Recreational places
    'tourism': ['hotel', 'museum', 'attraction'],  # Tourist locations
    'public_transport': True,  # Public transport infrastructure
    'railway': True  # Railway features
}

# Flatten the dictionary for querying OSM data
flattened_tags = {}
for key, value in tags.items():
    if isinstance(value, list):
        for v in value:
            flattened_tags[key] = v
    else:
        flattened_tags[key] = value

# Now you can use the `flattened_tags` dictionary in `geometries_from_bbox`
north = 47.5032
south = 47.4696
east  = 19.0839
west = 19.0276

# Query for building block polygon layer using popular OSM tags
gdf = ox.geometries_from_bbox(north, south, east, west, tags=flattened_tags)

# Display the first few results
print(gdf.head())
```
