````python
import ipyleaflet as ipl
from shapely.geometry import Point
import osmnx as ox
import json

# Define a bounding box for a larger area of Budapest
north = 47.4950
south = 47.4900
east = 19.1250
west = 19.1050

# Download a street network within the bounding box
G = ox.graph_from_bbox(north, south, east, west)

# Convert the graph to GeoDataFrames (nodes, edges)
node_gdf, edge_gdf = ox.graph_to_gdfs(G, nodes=True, edges=True)

# Calculate the center point of the graph
if len(node_gdf) == 1:
    center = node_gdf.geometry.iloc[0]
else:
    mean_lat = node_gdf.geometry.y.mean()
    mean_lon = node_gdf.geometry.x.mean()
    center = Point(mean_lon, mean_lat)

# Query for restaurants within the bounding box
tags = {'amenity': 'restaurant'}
restaurants_gdf = ox.geometries_from_bbox(north, south, east, west, tags=tags)

# Convert the edge GeoDataFrame to GeoJSON using GeoPandas
edge_geojson_str = edge_gdf.to_json()  # Use GeoPandas to convert to a GeoJSON string
restaurants_geojson_str = restaurants_gdf.to_json()  # Convert restaurants GeoDataFrame to GeoJSON string

# Convert the GeoJSON strings to dictionaries for ipyleaflet
edge_geojson = json.loads(edge_geojson_str)
restaurants_geojson = json.loads(restaurants_geojson_str)

# Create a map
m = ipl.Map(center=(center.y, center.x), zoom=12)

# Add edge network layer
edge_layer = ipl.GeoJSON(data=edge_geojson, name='Street Network')
m.add_layer(edge_layer)

# Function to display tooltip on click
def restaurant_click(event, feature, **kwargs):
    # Extract restaurant name or any relevant info
    restaurant_name = feature['properties'].get('name', 'Unknown')
    # Show tooltip with the restaurant name
    tooltip = f'Restaurant: {restaurant_name}'
    m.add_layer(ipl.Marker(location=(event['lat'], event['lng']), draggable=False, title=tooltip))

# Add restaurants layer with click event
restaurants_layer = ipl.GeoJSON(data=restaurants_geojson, name='Restaurants', style={'color': 'red'})
restaurants_layer.on_click(restaurant_click)
m.add_layer(restaurants_layer)

# Add layer control to toggle layers
layer_control = ipl.LayersControl()
m.add_control(layer_control)

# Display the map
m
````

````python
import ipyleaflet as ipl
from shapely.geometry import Point
import osmnx as ox
import json
from ipywidgets import HTML
import random

# Define a bounding box for a larger area of Budapest
north = 47.4950
south = 47.4900
east = 19.1250
west = 19.1050

# Download a street network within the bounding box
G = ox.graph_from_bbox(north, south, east, west)

# Convert the graph to GeoDataFrames (nodes, edges)
node_gdf, edge_gdf = ox.graph_to_gdfs(G, nodes=True, edges=True)

# Calculate the center point of the graph
mean_lat = node_gdf.geometry.y.mean()
mean_lon = node_gdf.geometry.x.mean()
center = Point(mean_lon, mean_lat)



# Query for restaurants within the bounding box
tags = {'amenity': 'restaurant'}
restaurants_gdf = ox.geometries_from_bbox(north, south, east, west, tags=tags)

# Save edge and restaurant GeoDataFrames as UTF-8 encoded GeoJSON files
edge_geojson_path = "./sample/edge_network.geojson"
restaurants_geojson_path = "./sample/restaurants.geojson"

# Write edge GeoJSON to file with UTF-8 encoding
with open(edge_geojson_path, 'w', encoding='utf-8') as edge_file:
    edge_file.write(edge_gdf.to_json())

# Write restaurant GeoJSON to file with UTF-8 encoding
with open(restaurants_geojson_path, 'w', encoding='utf-8') as restaurant_file:
    restaurant_file.write(restaurants_gdf.to_json())

# Now, read the GeoJSON files back in UTF-8 encoding
with open(edge_geojson_path, 'r', encoding='utf-8') as edge_file:
    edge_geojson = json.load(edge_file)

with open(restaurants_geojson_path, 'r', encoding='utf-8') as restaurant_file:
    restaurants_geojson = json.load(restaurant_file)

# Create a map
m = ipl.Map(center=(center.y, center.x), zoom=12)

# Add layer control to toggle layers
layer_control = ipl.LayersControl(position='topright')
m.add(layer_control)

# Function to create random color
def random_color(feature):
    return {
        'color': 'black',
        'fillColor': random.choice(['red', 'yellow', 'green', 'orange']),
    }

# Add edge network layer
edge_layer = ipl.GeoJSON(
    data=edge_geojson, 
    name='Street Network', 
    hover_style={'color': 'blue', 'dashArray': '0', 'fillOpacity': 0.5},
    style_callback=random_color
)

# Define a style for the restaurants
restaurant_style = {
    'color': 'blue',
    'weight': 2,
    'opacity': 0.8,
    'fillColor': 'blue',
    'fillOpacity': 0.5,
}

# Add restaurants layer with tooltips
restaurants_layer = ipl.GeoJSON(
    data=restaurants_geojson,
    name='Restaurants',
    style=restaurant_style,  # Set default style
)

# Function to create the popup content for an edge
def create_edge_popup(feature):
    name = feature['properties'].get('name', 'Unknown')
    address = feature['properties'].get('address', 'No address available')
    return f"<strong>{name}</strong><br>{address}"

# Function to handle edge clicks
def on_edge_click(feature, **kwargs):
    coords = feature['geometry']['coordinates']  # Get the clicked coordinates
    lng, lat = coords[0][0], coords[0][1]  # Note: coords are (lon, lat)
    # Create and show a popup
    popup_content = create_edge_popup(feature)
    # Create an HTML widget for the popup content
    popup_message = HTML()
    popup_message.value = popup_content
    
    # Create the popup using ipyleaflet's Popup object
    popup = ipl.Popup(
        location=(lat, lng),  # Popup location (lat, lng)
        child=popup_message,  # Pass the HTML content
        close_button=True,  # Add a close button
        auto_close=False,  # Keep it open until manually closed
        close_on_escape_key=True  # Allow closing with the escape key
    )
    
    # Add the popup to the map
    m.add(popup)

# Function to create the popup content for a restaurant
def create_restaurant_popup(feature):
    name = feature['properties'].get('name', 'Unknown')
    address = feature['properties'].get('address', 'No address available')
    return f"<strong>{name}</strong><br>{address}"

# Function to handle restaurant clicks
def on_restaurant_click(feature, **kwargs):
    coords = feature['geometry']['coordinates']  # Get the clicked coordinates
    lng, lat = coords[0], coords[1]  # Note: coords are (lon, lat)
    # Create and show a popup
    popup_content = create_restaurant_popup(feature)
    popup_message = HTML()
    popup_message.value = popup_content
    
    # Create the popup using ipyleaflet's Popup object
    popup = ipl.Popup(
        location=(lat, lng),  # Popup location (lat, lng)
        child=popup_message,  # Pass the HTML content
        close_button=True,  # Add a close button
        auto_close=False,  # Keep it open until manually closed
        close_on_escape_key=True  # Allow closing with the escape key
    )
    
    # Add the popup to the map
    m.add(popup)

# Connect the click event to the restaurant layer
restaurants_layer.on_click(on_restaurant_click)

edge_layer.on_click(on_edge_click)



# Add layers to the map
m.add(restaurants_layer)
m.add(edge_layer)

# Display the map
m
````
